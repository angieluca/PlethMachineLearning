```python
# Loading Data
data_train = np.load('data_train_type2.npy', allow_pickle=True)
labels_train = np.load('labels_train_type2.npy', allow_pickle=True)

print(data_train.shape, labels_train.shape)
```

```python
# Counting number samples per class
vals, counts = np.unique(labels_train, return_counts=True)

plt.bar(vals, counts)
plt.xticks(range(7),range(7))
plt.xlabel('Classes',size=20)
plt.ylabel('# Samples per Class', size=20)
plt.title('Training Data (Total = '+str(data_train.shape[0])+' samples)',size=15);
```

```python
column_names = ['Ti (msec)', 'Te (msec)', 'PIF', 'PEF', 'TV', 'EV', 'RT (msec)', 'MV', 'P (msec)', 'f (bpm)', 'EIP (msec)', 'EEP (msec)', 'Penh', 'EF50', 'RH', 'Tbox', 'VCF', 'AV', 'Sr']
```

```python
# Compute mean of each feature
feature_means = np.mean(data_train, axis=0)

# Plot the averages of the features
plt.figure(figsize=(10, 6))
plt.barh(column_names, feature_means, color='skyblue')
plt.xlabel('Average Value')
plt.ylabel('Features')
plt.title('Average Value of Features')
plt.show()
```

```python
# Compute mean and standard deviation of each feature
feature_means = np.mean(data_train, axis=0)
feature_std = np.std(data_train, axis=0)

# Plot the means with error bars representing standard deviation
plt.figure(figsize=(12, 8))
plt.bar(column_names, feature_means, yerr=feature_std, capsize=5, color='skyblue', alpha=0.7)
plt.xlabel('Features')
plt.ylabel('Mean Value')
plt.title('ANOVA Mean Value of Features with Standard Deviation Error Bars')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Print out the mean values
for i, feature in enumerate(column_names):
    print(f'{feature}: {feature_means[i]} ± {feature_std[i]}')
```

```python
column_names = ['Ti (msec)', 'Te (msec)', 'PIF', 'PEF', 'TV', 'EV', 'RT (msec)', 'MV', 'P (msec)', 'f (bpm)', 'EIP (msec)', 'EEP (msec)', 'Penh', 'EF50', 'RH', 'Tbox', 'VCF']

# Remove 'AV' feature from data
data_train = data_train[:, :-2]

# Compute mean and standard deviation of each feature
feature_means = np.mean(data_train, axis=0)
feature_std = np.std(data_train, axis=0)

# Plot the means with error bars representing standard deviation
plt.figure(figsize=(12, 8))
plt.bar(column_names, feature_means, yerr=feature_std, capsize=5, color='skyblue', alpha=0.7)
plt.xlabel('Features')
plt.ylabel('Mean Value')
plt.title('ANOVA Mean Value of Features with Standard Deviation Error Bars')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Print out the mean values
for i, feature in enumerate(column_names):
    print(f'{feature}: {feature_means[i]} ± {feature_std[i]}')
```

```python
# Perform ANOVA for each feature
anova_results = {}
for i, feature in enumerate(column_names):
    category_labels = labels_train.astype(str)  # Convert labels to strings if they are not
    unique_labels = np.unique(category_labels)
    samples = [data_train[category_labels == label, i] for label in unique_labels]
    anova_result = stats.f_oneway(*samples)
    anova_results[feature] = anova_result

# Visualize ANOVA results
p_values = [result.pvalue for result in anova_results.values()]
plt.figure(figsize=(10, 6))
plt.barh(column_names, p_values, color='skyblue')
plt.xlabel('P-value')
plt.ylabel('Features')
plt.title('ANOVA P-values for Features')
plt.show()
```
